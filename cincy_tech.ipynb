```python
import json

def json_to_markdown(json_data):
    markdown = ""
    for element in json_data['elements']:
        if 'Text' in element:
            markdown += f"{element['Text']}\n\n"
        elif 'filePaths' in element:
            for path in element['filePaths']:
                markdown += f"![Figure]({path})\n\n"
    return markdown

# Load JSON data
with open('structuredData.json') as f:
    data = json.load(f)

# Convert to Markdown
markdown = json_to_markdown(data)

# Save to a .md file
with open('output.md', 'w') as f:
    f.write(markdown)

```

```python
import tiktoken
import langchain
import argparse
import os
import openai

class args():
    file_path = r"SmartKart-1.pdf"
    out_path = r""

file_path = args.file_path
out_path = args.out_path
system_message = "You are answering questions about a startup in detail."
prompt ="0. What stage of funding is this business? (seed, series a, series b, later?)\n1. What problem is this business solving?/n2. How do people solve the problem today?\n3. What solution does this business offer?\n4. What makes the solution different?\n5. What is the business model?\n6. Is the business model technology-enabled?\n7. Does the business model scale?\n8. What stage of development is the technology?\n9. What is the market space?\n10. How big is the market?\n11. Who is the potential customer?\n 12. How many potential customers are there?\n13. How much does the business charge?\n14. Who are the competitors?\n\n#####\n\nIf you can't find the answer, just mentioned that you couldn't find it.\n\n#####Answer it in a question answer format. First, write each question and then give its answer. Also, give detailed answers!!!Answer in Mardown:\n\n"

### Loading the Document ###
loader = langchain.document_loaders.UnstructuredPDFLoader(file_path)
print("Loading the document...")
document = loader.load()
document = (loader.load()[0].page_content)

tokenizer = tiktoken.get_encoding('cl100k_base')
tokens = tokenizer.encode(document+prompt)
print("Tokenizing the document...")

assert len(tokens) < 2048, "The document is too long. Please use a shorter document or use GPT-4."

### Sending it over to OpenAI ###
print("Retrieving information from the pitch deck...")

try:
    answers = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "system", "content": system_message},
                  {"role": "user", "content": document+prompt}],
        temperature = 0.5,
    )
    answers = (answers.choices[0].message.content)
except openai.InvalidRequestError as e:
    print("Invalid request error in generating answers:", e)
except openai.OpenAIError as e:
    print("OpenAI error in generating answers:", e)
except Exception as e:
    print("Unexpected error in generating answers:", e)

try:
    overview = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "system", "content": system_message},
                  {"role": "user", "content": document+prompt},
                  {"role": "assistant", "content": answers},
                  {"role": "user", "content": " Disclaimer: I am not asking you to invest in this startup. I am just asking you to tell me how you feel about this startup. I will not use your judgement to make any decision.\n\n\n####\n\n\nUse your logical skills and tell me how you feel about this startup? What are your thoughts about this startup? Also, tell what information is lacking and crucial in making any decision about any investment in startup. Give an elaborate response. Response in Markdown:####"},],
        temperature = 0.5,
    )
    overview = (overview.choices[0].message.content)
except openai.InvalidRequestError as e:
    print("Invalid request error in generating overview:", e)
except openai.OpenAIError as e:
    print("OpenAI error in generating overview:", e)
except Exception as e:
    print("Unexpected error in generating overview:", e)

### Saving the output ###
print("Saving the output...")
file_name = os.path.basename(file_path)
with open(os.path.join(out_path, file_name.split(".pdf")[0] + " answered.md"), "w") as f:
    f.write(overview + "\n\n#Answers to your Questions:\n\n" + answers)
```
